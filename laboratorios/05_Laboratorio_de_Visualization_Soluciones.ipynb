## Ejercicio:

Crea el gráfico inferior donde hay 2 ejes. Uno es el histograma de reales, y el otro el histograma acumuluado de reales.

Pon el acumulado en azul y sin transparencia, y el no acumulado en rojo pero con transparencia al 0.3.

Añade las leyendas y asegúrate de que estén en la misma zona que las de la solución.

```python
f, ax = plt.subplots()
ax.hist(reales,cumulative=True, color = 'blue', label = 'acumulado')
ax2 = ax.twinx()
ax2.hist(reales, color = 'red', alpha = 0.3, label = 'no acumulado')
ax.legend(loc = 'upper right')
ax2.legend(loc = 'upper left');
```

## Ejercicio:

Crea un gráfico de barras de la variable smoker como el de la solución.

```python
smoker = df.smoker

f, ax = plt.subplots()

x, y = np.unique(smoker,return_counts = True)

ax.bar(x,y)
ax.set_title('Número de fumadores')
ax.set_xlabel('Fumador o no')
ax.set_ylabel('Frecuencia');
```

## Ejercicio:

Crea un gráfico de barras de la variable day, pero poniendo las etiquetas del eje x a tamaño 16 y rotandolas 90 grados.

```python
conteo = df.day.value_counts().sort_index()

f, ax = plt.subplots()
ax.bar(conteo.index,conteo.values)
ax.tick_params(axis = 'x', labelsize = 16, labelrotation = 90)
;
```

## Ejercicio:

Representa en el mismo gráfico de líneas el total de la factura y de las propinas en un gráfico que mida 16x6.

```python
f, ax = plt.subplots(figsize = (16,6))

ax.plot('total_bill', data = df)
ax.plot('tip', data = df);
```

## Ejercicio:

Con el mismo gráfico que antes restringe el eje x entre los valores 100 y 150 para ver un zoom.

```python
f, ax = plt.subplots(figsize = (16,6))

ax.plot('total_bill', data = df)
ax.plot('tip', data = df)
ax.set_xlim(100,150);
```

## Ejercicio:

Con el mismo gráfico que antes haz que la línea del total de la factura sea punteada, y que la línea de las propinas sea 10 veces más gruesa.

```python
f, ax = plt.subplots(figsize = (16,6))

ax.plot('total_bill', data = df, ls = ':')
ax.plot('tip', data = df, lw = 10)
ax.set_xlim(100,150);
```
